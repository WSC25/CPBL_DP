## 計算球隊雙殺打比例

```python
# Calculate the sum of '打席' for each '客場'
AWAY_PA = AWAYBOX.groupby('客場')['打席','雙殺', '打數'].sum().reset_index()

# Print the calculation result
AWAY_PA[['客場', '打席','雙殺','打數']]
AWAY_PA
```

```python
HOME_PA = HOMEBOX.groupby('主場')['打席','雙殺', '打數'].sum().reset_index()

# Print the calculation result
HOME_PA[['主場', '打席','雙殺','打數']]
HOME_PA
```

```python
away_df = pd.DataFrame(AWAY_PA)
home_df = pd.DataFrame(HOME_PA)

# Merge the two DataFrames based on the team name (球隊名)
merged_df = pd.merge(away_df, home_df, left_on='客場', right_on='主場', suffixes=('_客場', '_主場'))

# Drop the duplicate team name column
merged_df.drop('主場', axis=1, inplace=True)

# Calculate the total for "打席", "雙殺", and "打數" for each team
merged_df['打席總數'] = merged_df['打席_客場'] + merged_df['打席_主場']
merged_df['雙殺總數'] = merged_df['雙殺_客場'] + merged_df['雙殺_主場']
merged_df['打數總數'] = merged_df['打數_客場'] + merged_df['打數_主場']

# Drop the individual columns for away and home data
merged_df.drop(['打席_客場', '打席_主場', '雙殺_客場', '雙殺_主場', '打數_客場', '打數_主場'], axis=1, inplace=True)

merged_df.rename(columns={'客場': '球隊'}, inplace=True)

merged_df
```

```python
merged_df['雙殺打比例'] = merged_df['雙殺總數'] / merged_df['打數總數']
sorted(merged_df)
merged_df
```

```python
merged_df.to_excel('CPBL_球隊DP.xlsx', index=False)
```
